```python
pick_x_range = np.linspace(0.4 , 0.7 , 6)
pick_y_range = np.linspace(-0.3 , 0 , 6)
fixed_z = 0.215
grid_1 , grid_2 = np.meshgrid(pick_x_range, pick_y_range)
pick_points = np.array(list(zip(grid_1.flatten() , grid_2.flatten() , [fixed_z] * len(grid_1.flatten()))))
```

```python
pick_points[0]
```

```python
pick_points[0]
```
